```python
import customtkinter as ctk
import pywhatkit as pwk

# Appearance settings
ctk.set_appearance_mode("dark")
ctk.set_default_color_theme("blue")

# Main app window
app = ctk.CTk()
app.geometry("400x300")
app.title("My Page")

# --- Widgets ---
label1 = ctk.CTkLabel(app, text="Welcome to my app!")
label1.pack(pady=10)

search_entry = ctk.CTkEntry(app, placeholder_text="Search here")
search_entry.pack(pady=10)

output_box = ctk.CTkTextbox(app, height=100, width=350)
output_box.pack(pady=10)

# --- Functions ---
def do_search():
    query = search_entry.get()
    if query.strip():
        pwk.search(query)  # Google search in browser
        output_box.insert("end", f"Searched for: {query}\n")

def close_app():
    app.destroy()

# --- Buttons ---
search_button = ctk.CTkButton(app, text="Search with pwk", command=do_search)
search_button.pack(pady=5)

close_button = ctk.CTkButton(app, text="Close", command=close_app)
close_button.pack(pady=5)

# Run app
app.mainloop()
```

```python
import customtkinter as ctk

# Appearance settings
ctk.set_appearance_mode("dark")
ctk.set_default_color_theme("blue")

# Main app window
app = ctk.CTk()
app.geometry("400x400")
app.title("To-Do List App")

# --- Widgets ---
title_label = ctk.CTkLabel(app, text="My To-Do List", font=("Arial", 20))
title_label.pack(pady=10)

task_entry = ctk.CTkEntry(app, placeholder_text="Enter a new task")
task_entry.pack(pady=10)

task_listbox = ctk.CTkTextbox(app, height=200, width=350)
task_listbox.pack(pady=10)

# --- Functions ---
def add_task():
    task = task_entry.get()
    if task.strip():
        task_listbox.insert("end", f"â€¢ {task}\n")
        task_entry.delete(0, "end")

def clear_tasks():
    task_listbox.delete("1.0", "end")

def close_app():
    app.destroy()

# --- Buttons ---
add_button = ctk.CTkButton(app, text="Add Task", command=add_task)
add_button.pack(pady=5)

clear_button = ctk.CTkButton(app, text="Clear All", command=clear_tasks)
clear_button.pack(pady=5)

close_button = ctk.CTkButton(app, text="Close", command=close_app)
close_button.pack(pady=5)

# Run app
app.mainloop()
```
