```python
import sqlite3 as sql
conn= sql.connect("yogesh_info.db")
mycursor=conn.cursor()
```

```python
query="create table yoo(username char(255)unique,password char(255))"
mycursor.execute(query)
conn.commit()
```

```python
query2="select * from yoo"
mycursor.execute(query2)
data=mycursor.fetchall()
print(data)
```
